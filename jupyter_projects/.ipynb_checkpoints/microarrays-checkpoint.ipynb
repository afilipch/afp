## Microarray analyses for C.glutamicum

This analyses aims for the following:

* find the groups of genes which cooperatively change their expression in a wide range of conditions
* find the groups of experiments with similar gene expression patterns
* find blocks of genes which behave similarly in particular sets of experiments

## Experiments basic statistics:

```python
def draw_stat_intensity_histogram(df, func, axis_num, label, stat_label, percentile, bins):
    values = df.apply(func, axis=axis_num)
    resolution_limit = np.percentile(values, percentile)
    resolution = [x for x in values if x <= resolution_limit]

    fig, axes = plt.subplots(1,2, figsize=(12,4))
    fig.suptitle("%s [%s]" % (label, stat_label))
    for ax, mylist, title in zip(axes, [values, resolution], ['all', 'lower %d%%' % percentile]):
        ax.set_ylabel("Number of %s" % label)
        ax.set_xlabel("%s [%s]" % (stat_label, title))
        ax.hist(mylist, bins = bins);
        
def count_zeros(mylist):
    return len([x for x in mylist if x == 0])/(len(mylist))



```

```python
draw_stat_intensity_histogram(df, count_zeros, 1, "Experiments", "Fraction of genes with no intensity", 90, 25)
```

Interpretation: 
we observe a significant fraction of experiments with zero expression for more than 50% percent of genes. Further these experiments will be deleted from the initial table to ensure better performance of clustering algorithms

```python
draw_stat_intensity_histogram(df, np.mean, 1, "Experiments", "Mean", 95, 25)
```

Interpretation: 
we observe a number of experiments with abnormal (far away from average) total intensity. Further the experiments will be deleted to avoid potential biases.

```python
draw_stat_intensity_histogram(df, scs.variation, 1, "Experiments", "Coefficient of variation", 90, 25)
```

Interpretation: 
Experiments show pretty wide range of gene expression variation. However we will not delete outliers, since they may reflect strong experimental perturbations.

## Data Clearance
Here we remove genes and experiments which may disturb/bias the downstream analyses

```python
FRACTION_ZEROS = 0.25
def filter_fraction_zeros(mylist):
    return count_zeros(mylist) <= FRACTION_ZEROS

central_df = df.loc[:, df.apply(filter_fraction_zeros, axis=0)]
central_df = central_df.loc[central_df.apply(filter_fraction_zeros, axis=1)]

print("Filtering by nonzero fraction (%d%%):\nNumber of experiments:\t%d(%d)\nNumber of genes:\t%d(%d)" 
      % ((1- FRACTION_ZEROS)*100, central_df.shape[0], df.shape[0], central_df.shape[1], df.shape[1]))
```

```python
GENES_MEAN_RANGE = 0.8, 1000000
def filter_genes_mean(mylist):
    mymean = np.mean(mylist)
    return(mymean >= GENES_MEAN_RANGE[0] and mymean <= GENES_MEAN_RANGE[1])


EXP_MEAN_RANGE = 0.6, 2
def filter_exp_mean(mylist):
    mymean = np.mean(mylist)
    return(EXP_MEAN_RANGE[0] and mymean <= EXP_MEAN_RANGE[1])

central_df = central_df.loc[:, central_df.apply(filter_genes_mean, axis=0)]
central_df =  central_df.loc[central_df.apply(filter_exp_mean, axis=1)]
to_format = list(GENES_MEAN_RANGE) + list(EXP_MEAN_RANGE) + [central_df.shape[0], df.shape[0], central_df.shape[1], df.shape[1]]
print("Filtering by abnormal intensity (genes: %1.1f~%1.1f; experiments: %1.1f~%1.1f):\nNumber of experiments:\t%d(%d)\nNumber of genes:\t%d(%d)" 
      % tuple(to_format))

```

## We split the DataFrame into:
* microarray intensities (DataFrame values) as numpy array
* experiment names (DataFrame rows) as list
* gene names (DataFrame columns) as list

```python
data = central_df.to_numpy()
cond_names = list(central_df.index.values)
gene_names = list(central_df.columns.values)
```

```python
def normalize_to_sum(l):
    norma = sum(l)
    return [x/norma for x in l]

data_normed = np.apply_along_axis( normalize_to_sum, axis=1, arr=data )


def normalize_to_nonzero_zscores(l):
    nonzero = [x for x in l if x != 0]
    st = np.std(nonzero)
    mean = np.mean(nonzero)
    return [(x-mean)/st if x!=0 else x for x in l]

data_normed = np.apply_along_axis( normalize_to_nonzero_zscores, axis=0, arr=data_normed )
```

## TSNE analyses
TSNE is a simple way to represent multidimensional data into 2D map. This is done to visually inspect the cluster structure of the data. Please find more here: https://www.displayr.com/using-t-sne-to-visualize-data-before-prediction/

```python
model = TSNE(learning_rate=200)

# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(data_normed)
plt.title("TSNE for experiments")
plt.scatter(tsne_features[:,0], tsne_features[:,1]);
```

Interpretation:
Comprehension of the multidimensional expression data (expression of each gene is a dimension for an experiment) does not show any obvious cluster. However it does not mean, that there are no clusters for the experiments.

```python
model = TSNE(learning_rate=200)

# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(np.transpose(data_normed))
plt.title("TSNE for genes")
plt.scatter(tsne_features[:,0], tsne_features[:,1]);
```

Interpretation:
Comprehension of the multidimensional expression data (expression in each experiment is a dimension for a gene) shows several small-size clusters. It gives a promise that groups of genes with similar expression patterns can be detected.

## Hierarchical Clustering
Hierarchical Clustering is an interpretable method to find groups of similar objects (in our case: genes and experiments). Please find more here: https://www.displayr.com/what-is-hierarchical-clustering/

```python
def significant_changes(l):
    return [int(x/2) for x in l]

data_for_histogram_conditions = np.apply_along_axis( significant_changes, axis=0, arr=data_normed )
mergings = linkage(data_for_histogram_conditions, method='complete')
plt.title("Experiment based hierarchical clustering")
dendrogram(mergings, leaf_rotation=90, leaf_font_size=6);
```

```python
def represent_mergings(mergings, labels, type_, top=100, minsize=2):
    number2labels = {};
    start = len(labels);
    for n1, n2, dist, size in mergings[:top]:
        numbers = []
        for el in (n1, n2):
            n = int(el)
            if(n in number2labels):
                numbers.extend(number2labels[n])
            else:
                numbers.append(n)
        number2labels[start] = list(numbers)
        start+=1;
        #print(numbers, size)
        if(size<minsize):
            continue        
        if(type_=='condition'):
            print("Conditions\tdistance")
            for x in numbers:
                print("%s\t%1.3f" % (labels[x], dist) )
            print()
        if(type_=='gene'):
            print("%s\t%1.3f" % (", ".join([labels[x] for x in numbers]), dist) )
            
    
    
    
```

## Table of similar Experiments

```python
represent_mergings(mergings, cond_names, type_='condition', top=20)
```

## Genes Hierarchical Clustering

```python
def significant_changes_abs(l):
    return [abs(int(x/2)) for x in l]

data_for_histogram_genes = np.apply_along_axis( significant_changes_abs, axis=0, arr=data_normed )
data_for_histogram_genes = np.transpose(data_for_histogram_genes)
mergings = linkage(data_for_histogram_genes, method='complete')
plt.title("Genes based hierarchical clustering")
dendrogram(mergings, leaf_rotation=90, leaf_font_size=6);
```

## Table of similar genes

```python
print("Genes\tdistance")
represent_mergings(mergings, gene_names, type_='gene', top=200, minsize = 4)
```
